```python
import pandas as pd
from datetime import datetime
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
df=pd.read_csv('data/sphist.csv')#parse_dates=['Date']
df.head()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python

date_to_compare=datetime(year=2015,month=4,day=1)
date_to_compare
```

```python
df.sort_values(by='Date',inplace=True)
```

```python
df.head()
```

```python
#df.set_index('Date',inplace=True)
df['5day_avg']=df['Close'].rolling(5).mean().shift(1)


df['30day_avg']=df['Close'].rolling(30).mean().shift(1)
df['365day_avg']=df['Close'].rolling(365).mean().shift(1)

df['5day_std']=df['Close'].rolling(5).std().shift(1)
#df['30day_std']=df['Close'].rolling(30).std().shift(1)
df['365day_std']=df['Close'].rolling(365).std().shift(1)


        
    
```

```python
df.head(35)
```

```python
# Calculating complex indicators involving 3 primary indicators
df["ratio_avg_5_to_365"] = df["5day_avg"]/df["365day_avg"] 
df["ratio_std_5_to_365"] = df["5day_std"]/df["365day_std"]   
```

```python
df.tail(35)
```

Since you're computing indicators that use historical data, there are some rows where there isn't enough historical data to generate them. Some of the indicators use 365 days of historical data and the dataset starts on 1950-01-03. Thus, any rows that fall before 1951-01-03 don't have enough historical data to compute all the indicators. You'll need to remove these rows before you split the data.

```python
df=df[df['Date']>datetime(year=1951,month=1,day=2)].copy()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
train=df[df['Date']<datetime(year=2013,month=1,day=1)].copy()
test=df[df['Date']>=datetime(year=2013,month=1,day=1)].copy()
```

```python
train.info()
```

```python
test.info()
```

```python
lr=LinearRegression()
features=train.columns[7:]
target='Close'
lr.fit(train[features],train[target])
predictions=lr.predict(test[features])
test['Predicted_Close']=predictions
mae=mean_absolute_error(predictions,test[target])
```

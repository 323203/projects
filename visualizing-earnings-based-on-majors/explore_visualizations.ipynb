# Pandas, Histograms


```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(0,4):
    ax = fig.add_subplot(4,1,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```

```python
recent_grads['Sample_size'].hist(range=(0,5000))
plt.show()
```

```python
recent_grads['Unemployment_rate'].hist(bins=40)
plt.show()
```

```python
recent_grads['Men'].hist(bins=20)
plt.show()
```

```python
recent_grads['Women'].hist(bins=20)
plt.show()
```

# Pandas, Scatter Matrix Plot
